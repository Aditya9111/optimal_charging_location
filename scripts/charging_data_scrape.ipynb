## Scrape Data from GOOGLE MAPS API by Keyword like Restaurant, Shopping Malls , Food Court , Apartment etc

```python
def miles_to_meters(miles):
    try:
        return miles * 1_609.344
    except:
        return 0
        
API_KEY = 'API_KEY'
map_client = googlemaps.Client(API_KEY)

address = 'Bangalore'
geocode = map_client.geocode(address=address)
(lat, lng) = map(geocode[0]['geometry']['location'].get, ('lat', 'lng'))


search_string = input('Enter the Keyword you want to search in places:')
distance = miles_to_meters(100)
business_list = []

response = map_client.places_nearby(
    location=(lat, lng),
    keyword=search_string,
    radius=distance
)   

business_list.extend(response.get('results'))
next_page_token = response.get('next_page_token')

while next_page_token:
    time.sleep(2)
    response = map_client.places_nearby(
        location=(lat, lng),
        keyword=search_string,
        radius=distance,
        page_token=next_page_token
    )   
    business_list.extend(response.get('results'))
    next_page_token = response.get('next_page_token')


output_csv =[]
for data in business_list:
  output_csv.append({'Name':data['name'] ,'ID':data['place_id'] ,'Vicinity':data['vicinity'] ,'latitude':data["geometry"]["location"]["lat"] ,'longitude':data["geometry"]["location"]["lng"] ,'Rating':data['rating'] ,'Url':'https://www.google.com/maps/place/?q=place_id:' + data['place_id']  })


print(business_list)

```

```python
df = pd.DataFrame(output_csv)
```
