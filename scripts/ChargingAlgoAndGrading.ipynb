```python
import pandas as pd
import csv
import numpy as np
```

```python
final_df = pd.read_csv('parking_analysed.csv')
final_df.head()
```

```python
def charging_algo(final_df):
  final_df['min'] = final_df['charging_station_nearby'] + final_df['FREQ']*10
  final_df['rest'] = final_df['Food_court_nearby']*5 + final_df['Gas_station_nearby']*10 + final_df['Malls_nearby']*5 + final_df['Apartments_nearby']*10 
  final_df['final_point'] = final_df['min'] + final_df['rest']

charging_algo(final_df)
```

```python
final_df = final_df.sort_values(by=['final_point'])
```

```python
minvalue = final_df['final_point'].min()
maxvalue = final_df['final_point'].max()
  
minvalue , maxvalue
```

## GRADE

```python
range = 1000
```

```python
def grade_station(dl):
    if minvalue <= dl <= minvalue + range: return 10
    elif minvalue + range < dl <= minvalue + 1.5*range: return 9
    elif minvalue + 1.5*range < dl <= minvalue + 1.8*range: return 8
    elif minvalue + 1.8*range < dl <= minvalue + 2*range: return 7
    elif minvalue + 2*range < dl <= minvalue + 2.2*range: return 6
    elif minvalue + 2.2*range < dl <= minvalue + 2.4*range: return 5
    elif minvalue + 2.4*range < dl <= minvalue + 2.5*range: return 4
    elif minvalue + 2.5*range < dl <= minvalue + 2.6*range: return 3
    elif minvalue + 2.6*range < dl <= minvalue + 2.7*range: return 2
    elif minvalue + 2.7*range < dl <= minvalue + 3*range: return 1
    else: return 'None'

final_df["grade"] = final_df['final_point'].map(grade_station)
```
