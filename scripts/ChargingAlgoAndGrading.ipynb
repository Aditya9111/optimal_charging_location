```python
import pandas as pd
import csv
import numpy as np
```

```python
final_df = pd.read_csv('parking_analysed.csv')
final_df.head()
```

```python
def charging_algo(final_df):
  final_df['min'] = final_df['charging_station_nearby'] + final_df['FREQ']*10
  final_df['rest'] = final_df['Food_court_nearby']*5 + final_df['Gas_station_nearby']*10 + final_df['Malls_nearby']*5 + final_df['Apartments_nearby']*10 
  final_df['final_point'] = final_df['min'] + final_df['rest']

charging_algo(final_df)
```

```python
final_df.head()
```

```python
print("Mean:Computed column-wise:");

meanData = final_df['final_point'].mean();

print(meanData);

print("Median:Computed column-wise:");

medianData = final_df['final_point'].median();

print(medianData);

print("Mode:Computed column-wise:");

modeData = final_df['final_point'].mode();

print(modeData);
```

```python
final_df = final_df.sort_values(by=['final_point'])
```

```python
final_df.head()
# final_df.to_csv('final_point_result.csv')
```

```python
minvalue = final_df['final_point'].min()
maxvalue = final_df['final_point'].max()
  
minvalue , maxvalue
```

## GRADE

```python
def grade_station(dl):
    if minvalue <= dl <= minvalue+1000: return 1
    elif minvalue+1000 < dl <= minvalue+1500: return 2
    elif minvalue+1500 < dl <= minvalue+1800: return 3
    elif minvalue+1800 < dl <= minvalue+2000: return 4
    elif minvalue+2000 < dl <= minvalue+2200: return 5
    elif minvalue+2200 < dl <= minvalue+2400: return 6
    elif minvalue+2400 < dl <= minvalue+2600: return 7
    elif minvalue+2600 < dl <= minvalue+2800: return 8
    elif minvalue+2800 < dl <= minvalue+2900: return 9
    elif minvalue+2900 < dl <= minvalue+3000: return 10
    else: return 'None'

final_df["grade"] = final_df['final_point'].map(grade_station)
```
