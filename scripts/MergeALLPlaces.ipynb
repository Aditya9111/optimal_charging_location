## Merge all scraped data in a single dataframe ,so that we can run our also on that final dataframe 

```python
ap_nearby_df = pd.read_csv('no_of_apartment_nearby.csv')
cs_nearby_df = pd.read_csv('no_of_charging_nearby.csv')
food_nearby_df = pd.read_csv('no_of_food_court_nearby.csv')
gas_nearby_df = pd.read_csv('no_of_gas_nearby.csv')
malls_nearby_df = pd.read_csv('no_of_malls_nearby.csv')
```

```python
malls_nearby_df.head()
```

```python
first_merge = ap_nearby_df.merge(cs_nearby_df[['Name','charging_station_nearby','parking_id']])
second_merge = first_merge.merge(food_nearby_df[['Name','Food_court_nearby','parking_id']])
third_merge = second_merge.merge(gas_nearby_df[['Name','Gas_station_nearby','parking_id']])
final_merge = third_merge.merge(malls_nearby_df[['Name','Malls_nearby','parking_id']])
```

```python
final_merge.tail()
```

```python
final_merge.to_csv('final.csv')
```
